```python
import pandas as pd
import json
```

The data is split into 3 json files at the source - book_details contains the name of the book, the star rating distribution (from which I intend to get an overall rating), total review count, and publication date as columns for day, month, and year. The files containing reviews have the review text, the review date, and the reviewer's rating.

```python
# This is the initial look at the data. 
book_details = pd.read_json(r'D:\Programming\LHL-data-notes\Final\data\goodreads_book_works.json', lines=True)
book_details.head()
```

```python
romance_reviews = pd.read_json(r'D:\Programming\LHL-data-notes\Final\data\goodreads_reviews_romance.json', lines=True)
ya_reviews = pd.read_json(r'D:\Programming\LHL-data-notes\Final\data\goodreads_reviews_young_adult.json', lines=True)
romance_reviews.head()
```

```python
book_details = book_details.rename(columns={'best_book_id': 'book_id'})
```

```python
# I merge this into a single frame
romance_df = pd.merge(romance_reviews, book_details, on="book_id", how="left")
ya_df = pd.merge(ya_reviews, book_details, on="book_id", how="left")
```
